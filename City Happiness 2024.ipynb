**Dataset Features:**

* **City**: Name of the city.
* **Month**: The month in which the data is recorded.
* **Year**: The year in which the data is recorded.
* **Decibel_Level**: Average noise levels in decibels, indicating the auditory comfort of the citizens.
* **Traffic_Density**: Level of traffic density (Low, Medium, High, Very High), which might impact citizens' daily commute and stress levels.
* **Green_Space_Area**: Percentage of green spaces in the city, positively contributing to the mental well-being and relaxation of the inhabitants.
* **Air_Quality_Index**: Index measuring the quality of air, a crucial aspect affecting citizens' health and overall satisfaction.
* **Happiness_Score**: The average happiness score of the city (on a 1-10 scale), representing the subjective well-being of the population.
* **Cost_of_Living_Index**: Index measuring the cost of living in the city (relative to a reference city), which could impact the financial satisfaction of the citizens.
* **Healthcare_Index**: Index measuring the quality of healthcare in the city, an essential component of the population's well-being and contentment.

# Conclusions
---

The key findings are summarized as follows:

1. Air Quality and Decibel Levels are almost perfectly negatively correlated with the overall Happiness Score as well as with all variables with positively correlated with the Happiness Score such as Cost of Living, Healthcare, and Green Space Area.
    * A high air quality index indicates that the air quality in a specific area is relatively poor. 
    * A high decibel level index indicates louder, more frequent noise in a specific area.
    * The air quality index at lower values showed a postive relationship between air quality and happiness score, but the higher air quality index values have a widely spread out distribution. This could indicate that while air quality index doesn't directly relate to the happiness score, the happiest groups have on average a lower air quality index.

2. The regression models for Happiness Score by Air Quality and Happiness Score by Decibel Levels in regards to happiness groups look virtually identical to one another. 
    * Both have a wide distribution among the lowest group.
    * Both show less impact on overall happiness among the higher groups. 

3. Happiness Score has the strongest relationship with Healthcare Index and Green Space Area. 
    * A high healthcare index would suggest that a city has a well-developed and effective healthcare system, with good access to medical services, quality of care, and positive health outcomes for the population. While a low healthcare index may indicate challenges in healthcare accessibility, quality, or outcomes. 
    * The more green space a city has, the better the air quality and the lower the noise level. Better air quality results in better overall health which corresponds to a better overall happiness score. 

4. The cost of living strongly impacts Happiness Score overall, but midly in comparison to access to healthcare and the amount of green space within a city.

5. Happiness Score is most strongly influenced by the Healthcare Index 

6. There is a nuanced relationship between the Cost of Living Index and Happiness Scores.
    * Lower Cost of Living Index Values:
        - Having lower values of the Cost of Living Index doesn't necessarily mean high happiness scores.
        - Within the lowest group (the group with the lowest Cost of Living Index values), the average happiness score is lower.
    * Higher Cost of Living Values:
        - Higher happiness scores are observed in groups with higher Cost of Living values. This indicates that, on average, individuals or groups with a higher cost of living tend to have higher happiness scores.
